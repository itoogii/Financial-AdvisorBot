# Training Dataset

This notebook is created to experiment and work on historical stock price data. 

I start with converting the dataset (downloaded from Kaggle, author used yfinance) to parquet format. I am using this dataset for my personal and educational purposes.  

```python
# import data into pandas dataframe
import pandas as pd

def import_data(file_path: str) -> pd.DataFrame:
    try:
        data = pd.read_csv(file_path)
        return data
    except Exception as e:
        print(f"An error occurred while importing data: {e}")
        return pd.DataFrame()
```

```python
import yfinance as yf

def load_history(ticker: str, start_date: str, end_date: str) -> pd.DataFrame:
    try:
        stock = yf.Ticker(ticker)
        history = stock.history(
            start=start_date, 
            end=end_date, 
            interval="1d", 
            prepost=False, 
            auto_adjust = False) # auto_adjust added to get same data as in csv files
        return history
    except Exception as e:
        print(f"An error occurred while loading history: {e}")
        return pd.DataFrame()
```

The following code loops through the yfinance folder and convert them to the parquet format. The dataset from Kaggle contains data until April 2020, I downloaded last 5 years and 8 months of data for test dataset and saving them to parquet format as well. It skips the tickers that probably no longer exist. 

The following code is disabled (condition) on purpose to avoid running again. 

```python
if False: # adding the condition to avoid running this code again. 
    from pathlib import Path
    import time
    Path("parquet_train_data").mkdir(parents=True, exist_ok=True)
    Path("parquet_test_data").mkdir(parents=True, exist_ok=True)

    files = Path("yfinance_stock_data").glob("*.csv")
    for stock in files:
        ticker = stock.stem
        try:
            df_test = load_history(ticker, "2020-04-02", "2026-02-01")
            if df_test.empty:
                print(f"No data found for {ticker}. Skipping...")
                continue 
            if hasattr(df_test.index, 'tz') and df_test.index.tz is not None:
                df_test.index = df_test.index.tz_localize(None)
            df_test = df_test[['Open', 'High', 'Low', 'Close', 'Adj Close','Volume']]
            df_test.to_parquet(f"parquet_test_data/{ticker}.parquet", engine='pyarrow', compression='snappy')
            df_train = import_data(stock) 
            df_train['Date'] = pd.to_datetime(df_train['Date'])
            df_train.set_index('Date', inplace=True)
            df_train.to_parquet(f"parquet_train_data/{ticker}.parquet", engine='pyarrow', compression='snappy')
            time.sleep(1)  # To avoid rate limit on yfinance API
        except Exception as e:
            print(f"An error occurred while processing {ticker}: {e}")
```
