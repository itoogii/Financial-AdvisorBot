```python
from dotenv import load_dotenv
from crewai import LLM, Agent, Task, Crew
import os
import yaml

load_dotenv()
model = "azure/gpt-5.1-chat"
api_url = os.environ.get("AZURE_API_BASE")
api_key = os.environ.get("AZURE_API_KEY")
api_version = os.environ.get("AZURE_API_VERSION")

llm = LLM(
    model="azure/gpt-5.1-chat",
    endpoint=api_url,
    api_key=api_key,
    api_version=api_version,
)
```

### Loading tasks from yaml

```python
files={
    'agents': 'config/agents.yaml',
    'tasks': 'config/tasks.yaml'
}

configs={}
for config_type, file_path in files.items():
    with open(file_path, 'r') as file:
        configs[config_type] = yaml.safe_load(file)

agents_config = configs['agents']
tasks_config = configs['tasks']
```

```python
financial_advisor_agent = Agent(
    config=agents_config['financial_advisor_agent']
)
advisory_task = Task(
    config=tasks_config['advisory_task'],
    agent=financial_advisor_agent
)
```

```python
crew = Crew(
    agents=[financial_advisor_agent],
    tasks=[advisory_task],
    verbose=True
)
```

```python
result=crew.kickoff()
```
