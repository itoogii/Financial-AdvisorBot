#### yfinance experiment

```python
import yfinance as yf
import pandas as pd
from pathlib import Path
import ta
import tqdm
from itertools import accumulate
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


sns.set_style("darkgrid")
```

```python
ticker = "AAPL"
df = yf.download(tickers = ticker ,       # list of tickers
                  start = "2020-04-01", # start date
                  end = "2026-02-01",   # end date
                  interval = "1d",       # trading interval
                  ignore_tz = True,      # ignore timezone when aligning data from different exchanges?
                  prepost = False)
```

```python
df.head()
```

```python
df.columns = df.columns.droplevel(1)
df.head()
```

```python
apple_stock = yf.Ticker(ticker)
apple_history=apple_stock.history(start = "2020-04-02", # start date
                  end = "2026-02-01",   # end date
                  interval = "1d",       # trading interval
                  prepost = False,
                  auto_adjust = False
)
apple_history.tail()
```

```python
if apple_history.index.tz is not None:
    apple_history.index = apple_history.index.tz_localize(None)
```

```python
apple_history.tail()
```
