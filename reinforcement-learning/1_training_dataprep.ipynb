# Training Dataset

This notebook is created to experiment and work on historical stock price data. 

```python
# import data into pandas dataframe
import pandas as pd

def import_data(file_path: str) -> pd.DataFrame:
    # Imports data from ....
    try:
        data = pd.read_csv(file_path)
        return data
    except Exception as e:
        print(f"An error occurred while importing data: {e}")
        return pd.DataFrame()
```

The following code loops through the yfinance folder and convert them to the parquet format. 

```python
from pathlib import Path

Path("parquet_stock_data").mkdir(parents=True, exist_ok=True)

files = Path("yfinance_stock_data").glob("*.csv")
for stock in files:
    ticker = stock.stem
    df = import_data(stock) 
    df.to_parquet(f"parquet_stock_data/{ticker}.parquet", engine='pyarrow', compression='snappy')
```
